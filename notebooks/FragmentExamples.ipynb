```python
from chemsampler.samplers.smallworld.sampler import SmallWorldSampler
```

```python
ss = SmallWorldSampler()
```

```python
sampled = ss.sample(["CC(=O)OC1=CC=CC=C1C(=O)O"])
```

```python
from rdkit.Chem import Draw
from rdkit import Chem

mols = [Chem.MolFromSmiles(smi) for smi in sampled]
Draw.MolsToGridImage(mols[:10])
```
